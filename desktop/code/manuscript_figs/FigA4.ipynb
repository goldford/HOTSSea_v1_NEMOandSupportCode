Purpose: Plot temperature anoms from files created in the Fig08 notebook for the 1970 - 2018 period
            for supplemental manuscript figure
         
Data in: Nanoose anomalies in .nc files
         
Data out: fig


```python
import xarray as xr
import os
import cartopy as cp
import matplotlib.pyplot as plt
import numpy as np
import glob
import pandas as pd
import cmocean as cm

outpath1 = '..//..//data//eval//nanoose_clima//'

checkoutput = True
if checkoutput == True:
#     obs_salt_clim_allyr = xr.open_dataset(os.path.join(outpath1, 'obs_salinity_1970-2018-climatology.nc'))
    obs_temp_clim_allyr = xr.open_dataset(os.path.join(outpath1, 'Nanoose_' + 'obs_anom_temp_1970-2018.nc'))

fig, ax = plt.subplots(1,1, figsize=(8, 3), facecolor='w', edgecolor='k')

theplot = xr.plot.pcolormesh(obs_temp_clim_allyr.temperature, 
                             ax = ax, cmap = cm.cm.balance, 
                             vmin = -2, vmax = 2,
                             cbar_kwargs={'label': "Temperature Anomaly ($^\circ$C)"})

ax.plot()

ax.invert_yaxis()
ax.set_title('')
plt.suptitle('')
ax.set_ylabel('Depth (m)')
ax.set_xlabel('')

plt.tight_layout()
plt.savefig('../../figs/FigA4.png', dpi=350)
plt.savefig('../../figs/FigA4.pdf', dpi=350)
    
```
